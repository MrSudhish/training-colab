```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(42)

X=np.linspace(-3,3,100)
y= X**3 + X**2 - 5*X + 10 + np.random.randn(100)*4
```

```python
X
```

```python
y
```

```python
X = X.reshape(-1,1)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X,y,label="Data Points", color='blue')
plt.title("Non Linear Data")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
lin_model= LinearRegression()
lin_model.fit(X,y)
y_pred=lin_model.predict(X)

plt.figure(figsize=(8,5))
plt.scatter(X,y,label="Data", color='blue')
plt.plot(X,y_pred,color='red',linewidth=3, label="Linear Fit (Bad)")
plt.title("Linear Regression Fails Non Linear Data")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
ploy= PolynomialFeatures(degree=3)
X_poly=ploy.fit_transform(X)

poly_model=LinearRegression()
poly_model.fit(X_poly,y)
y_pred=poly= poly_model.predict(X_poly)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X,y,label="Data", color='blue')
plt.plot(X,y_pred,color='red',linewidth=3, label="Linear Fit (Bad)")
plt.title("Polynimial Regresion Fails Non Linear Data")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
